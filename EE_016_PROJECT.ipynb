```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv

Input = "housing.csv.zip"
df = pd.read_csv(Input)

print(df.head())
print(df.shape)
print(df.size)
print(df.dtypes)
print(df.columns)

value_columns = ["longitude", "latitude", "housing_median_age", "total_rooms", "total_bedrooms",
                 "population", "households", "median_income", "median_house_value"]

for column in value_columns:
    print ("\n", column)
    print ("mean:", np.mean(df[column]))
    print ("max:", np.max(df[column]))
    print ("min:", np.min(df[column]))
    print ("std:", np.std(df[column]))
```

# Graphs

total_bedrooms vs median_house_value doesnt really work because data is aggregate

```python
plt.figure(figsize=(8,8))
plt.scatter(df["total_bedrooms"], df["median_house_value"], alpha=0.3)
plt.title("total_bedrooms vs median_house_value")
plt.xlabel("total_bedrooms")
plt.ylabel("median_house_value")
plt.show()
```

bedrooms_per_households again doesn't really work

```python
df["bedrooms_per_households"] = df["total_bedrooms"] / df["households"]

plt.figure(figsize=(8,8))
plt.scatter(df["bedrooms_per_households"], df["median_house_value"], alpha=0.3)
plt.title("bedrooms_per_households vs median_house_value")
plt.xlabel("bedrooms_per_households")
plt.ylabel("median_house_value")
plt.show()
```

rooms_per_households v. median_house_value again doesn't really work.

```python
df["rooms_per_households"] = df["total_rooms"] / df["households"]

plt.figure(figsize=(8,8))
plt.scatter(df["rooms_per_households"], df["median_house_value"], alpha=0.3)
plt.title("rooms_per_households vs median_house_value")
plt.xlabel("rooms_per_households")
plt.ylabel("median_house_value")
plt.show()
```

median_income v. median_house_value is alright

```python
plt.figure(figsize=(8,8))
plt.scatter(df["median_income"], df["median_house_value"], alpha=0.3)
plt.title("median_income v. median_house_value")
plt.xlabel("median_income")
plt.ylabel("median_house_value")
plt.show()
```

median_house_value histogram looks good

```python
plt.hist(df["median_house_value"], bins= 50)
plt.title("median_house_value")
plt.xlabel("median_house_value")
plt.ylabel("idk")
plt.show()
```

Works great. coatal cities are pricey (ligher colors) inland is cheaper (darker colors).

```python
plt.figure(figsize=(8,8))
plt.scatter(df["longitude"], df["latitude"], alpha=0.3, c =df["median_house_value"])
plt.title("location v. median_house_value")
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.show()
```

# Correlation Analysis



## Correlation matrix

```python
corr_matrix = df[value_columns].corr()
corr_matrix
```

## Heatmap

```python
import seaborn as sns

plt.figure(figsize=(8,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of housing data")
plt.show()
```

## Pairwise Graphs

Dataset is reduced to median income, total rooms, and housing median age.



```python
filtered_df = df[["median_income", "total_rooms", "housing_median_age"]]
sns.pairplot(filtered_df)
plt.suptitle("Pairwise Scatter Plots of housing data", y=1.02)
plt.show()
```

# Regression Model

## Prep

```python
from sklearn.linear_model import LinearRegression

x = filtered_df
y = df["median_house_value"]

model = LinearRegression()
model.fit(x, y)

print("Intercept: ", model.intercept_)
print("Coefficients: ", model.coef_)
```
